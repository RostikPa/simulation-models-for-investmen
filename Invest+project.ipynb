```python
import numpy as np
import scipy.stats as sp
import pandas as pd
import matplotlib.pyplot as plt
import math
import random
from sklearn import linear_model
from datetime import datetime
from matplotlib import style
```

```python
data= pd.read_csv('C:/Users/Admin/Desktop/IN_P/printer.csv', sep = ';')
start_v = 40000
r = 0.23
```

```python
data
```

```python
def NPV(dt):
    i=0
    npv = 0
    for i in range(len(dt['Income'])):
        npv = npv + dt.loc[i,'Income']*((1-r)**dt.loc[i,'Years'])-dt.loc[i,'Cost']*((1-r)**dt.loc[i,'Years'])
    npv = npv-start_v*((1-r)**0)
    return npv

def IRR(dt):
    r_i=0
    for j in range(100):
        i=0
        npv = 0
        for i in range(len(dt['Income'])):
            npv= npv + dt.loc[i,'Income']*((1-r_i)**dt.loc[i,'Years'])-dt.loc[i,'Cost']*((1-r_i)**dt.loc[i,'Years'])
        npv = npv-start_v*((1-r_i)**0)
        if npv <= 0:
            break
        r_i=r_i+0.01
    return r_i
def PI(dt):
    i=0
    npv = 0
    pi=0
    for i in range(len(dt['Income'])):
        npv = npv + dt.loc[i,'Income']*((1-r)**dt.loc[i,'Years'])-dt.loc[i,'Cost']*((1-r)**dt.loc[i,'Years'])
    npv = npv-start_v*((1-r)**0)
    pi = npv/start_v
    return pi

def PBP(dt):
    i=0
    P = 0
    pbp=10
    for i in range(len(dt['Income'])):
        P = P + dt.loc[i,'Income']*((1-r)**dt.loc[i,'Years'])-dt.loc[i,'Cost']*((1-r)**dt.loc[i,'Years'])
        if P > start_v:
            pbp = dt.loc[i,'Years']
            break
    return pbp
    
print(NPV(data))
print(IRR(data))
print(PI(data))
print(PBP(data))
```

```python
data_inv_rat = pd.DataFrame(columns=['NPV','IRR', 'PI', 'PBP'])
data_inv_rat.loc[0,'NPV'] = NPV(data)                            
data_inv_rat.loc[0,'IRR'] = IRR(data)
data_inv_rat.loc[0,'PI'] = PI(data)
data_inv_rat.loc[0,'PBP'] = PBP(data)
data_inv_rat
data_r = pd.DataFrame(columns=['Years','Income','Cost'])
data_r['Years']=data['Years']
i=1
for i in range(1000):
    j =0
    for j in range(len(data_r)):
        data_r.loc[j,'Income'] = random.normalvariate(data.loc[j,'Income'],(data.loc[j,'Income']/3))
        data_r.loc[j,'Cost'] = random.normalvariate(data.loc[j,'Cost'],(data.loc[j,'Cost']/3))
    data_inv_rat.loc[i,'NPV'] = NPV(data_r)                            
    data_inv_rat.loc[i,'IRR'] = IRR(data_r)
    data_inv_rat.loc[i,'PI'] = PI(data_r)
    data_inv_rat.loc[i,'PBP'] = PBP(data_r)
data_inv_rat
```

```python
data_inv = pd.DataFrame(columns=['Rate','NPV','IRR', 'PI', 'PBP'])
data_inv['Rate'] = ['M(x)', 'D(x)','S(x)', 'CV(x)']
data_inv.loc[0, 'NPV'] = data_inv_rat['NPV'].mean()
data_inv.loc[0, 'IRR'] = data_inv_rat['IRR'].mean()
data_inv.loc[0, 'PI'] = data_inv_rat['PI'].mean()
data_inv.loc[0, 'PBP'] = data_inv_rat['PBP'].mean()

data_inv.loc[1, 'NPV'] = data_inv_rat['NPV'].std()**2
data_inv.loc[1, 'IRR'] = data_inv_rat['IRR'].std()**2
data_inv.loc[1, 'PI'] = data_inv_rat['PI'].std()**2
data_inv.loc[1, 'PBP'] = data_inv_rat['PBP'].std()**2

data_inv.loc[2, 'NPV'] = data_inv_rat['NPV'].std()
data_inv.loc[2, 'IRR'] = data_inv_rat['IRR'].std()
data_inv.loc[2, 'PI'] = data_inv_rat['PI'].std()
data_inv.loc[2, 'PBP'] = data_inv_rat['PBP'].std()

data_inv.loc[3, 'NPV'] = data_inv_rat['NPV'].std()/data_inv.loc[0, 'NPV']
data_inv.loc[3, 'IRR'] = data_inv_rat['IRR'].std()/data_inv.loc[0, 'IRR']
data_inv.loc[3, 'PI'] = data_inv_rat['PI'].std()/data_inv.loc[0, 'PI']
data_inv.loc[3, 'PBP'] = data_inv_rat['PBP'].std()/data_inv.loc[0, 'PBP']
data_inv
```
